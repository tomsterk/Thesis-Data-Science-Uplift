```python
import os
import glob
import pandas as pd

files = glob.glob(os.path.join(os.getcwd(), "Input_phase_2", "*.csv"))
print(f"Found {len(files)} files: {files}")

df_all = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)
df_unique = df_all.drop_duplicates(subset="customer_nk")

df_unique[["customer_nk", "original_incentive_name"]].to_csv(
    os.path.join(os.getcwd(), "Input_phase_2", "Upload_crm_phase_2_winback.csv"),
    index=False
)
print(f"Saved {len(df_unique)} customers to Upload_crm_phase_2_winback.csv")
print(f"Total rows before dedup: {len(df_all)}")
print(f"Unique customers:        {len(df_unique)}")
```
